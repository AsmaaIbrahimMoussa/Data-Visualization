# Pie Chart

## Step 1: Import Libraries 

```python
import matplotlib.pyplot as plt
```

## Step 2: Basic Pie Chart
We start by creating a simple pie chart using values only.

```python
sizes = [15, 30, 45, 10]
plt.pie(sizes)
plt.show()
```

## Step 3: Add Labels and Explode
Next, we add labels to each segment and use `explode` to emphasize a slice.

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, wedgeprops={"edgecolor": "black"})
plt.show()
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, wedgeprops={"edgecolor": "black"}, autopct='%1.1f%%')
plt.show()
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors=['pink', 'rosybrown', 'gray', 'skyblue']
#colors= ['#00ffff', '#13eac9', '#000080', '#add8e6']

plt.pie(sizes, labels=labels, wedgeprops={"edgecolor": "black"}, autopct='%1.1f%%', colors=colors)
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.2, 0]
plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

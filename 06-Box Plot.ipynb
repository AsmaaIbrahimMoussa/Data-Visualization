# Box Plot: Visualizing Distribution and Outliers
Box plots are useful for visualizing the distribution, central tendency, and spread of data, as well as detecting outliers.

## Step 1: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Load Dataset
We'll use the built-in `tips` dataset from Seaborn.

```python
data = sns.load_dataset('tips')
data.head()
```

## Step 3: Basic Box Plot
Let's plot a basic box plot showing the distribution of tips.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['tip'])
plt.title('Distribution of Tips')
plt.xlabel('Tip Amount ($)')
plt.grid(True)
plt.show()
```

## Step 4: Box Plot by Category
Now we'll group the box plot by day to compare distributions.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='tip',hue='sex', data=data, palette='Set1')
plt.title('Tip Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')
plt.grid(True)
plt.show()
```

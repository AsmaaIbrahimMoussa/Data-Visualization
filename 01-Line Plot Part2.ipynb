# Advanced Line Plot

## Step 1: Import Libraries 

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

## Step 2: Load the data

```python
data = pd.read_csv("Dev_Salary.csv")
```

```python
data.head()
```

## Step 3: Change Line plot (Marker, Linestyle, Color)

```python
#Median Dev salary vs age
Dev_age=data["Age"]
Dev_salary=data["All_Devs"]
plt.plot(Dev_age, Dev_salary, label='All Devs', marker='o', linestyle='--', color='r')

py_Dev_salary=data["Python"]
plt.plot(Dev_age, py_Dev_salary, label='Python Devs', marker='v', linestyle='-.', color='g')

plt.title('Median Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

```python
#Median Dev salary vs age
Dev_age=data["Age"]
Dev_salary=data["All_Devs"]
plt.plot(Dev_age, Dev_salary, label='All Devs', marker='*', linestyle='--', color='r')

py_Dev_salary=data["Python"]
plt.plot(Dev_age, py_Dev_salary, label='Python Devs', marker='^', linestyle='-.', color='g')

JS_Dev_salary=data["JavaScript"]
plt.plot(Dev_age, py_Dev_salary, label='JavaScript Devs', marker='o', linestyle=':', color='y')

plt.title('Median Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

## Fivethirtyeight

**Effects of 'fivethirtyeight' style:**

- Uses a light grey background.

- Has bold and clear fonts for better readability.

- Includes grid lines for better data interpretation.

- Uses thicker lines for plot elements like curves and bars.

- Provides a modern and publication-ready appearance.


```python
#Median Dev salary vs age using fivethirtyeight style
plt.style.use('fivethirtyeight')
Dev_age=data["Age"]
Dev_salary=data["All_Devs"]
plt.plot(Dev_age, Dev_salary, label='All Devs', marker='o', linestyle='--', color='r')

py_Dev_salary=data["Python"]
plt.plot(Dev_age, py_Dev_salary, label='Python Devs', marker='v', linestyle='-.', color='g')

plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.title('Median Salary vs Age')
plt.legend()
plt.show()

```

# Scatter Plot
Scatterplots are a commonly used data visualisation tool to allow us to identify and determine if there is a relationship between two variables. We will also be able to tell if that relationship is a strong one or if there is no relationship.

# Step 1: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Load data 

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

## Step 3: Simple Scatter plot
We plot `total_bill` against `tip` to see if there's a correlation.

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data["total_bill"], data["tip"])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

## Step 4: Enhance the Scatter Plot using Seaborn


we 'll use hue and style:
- hue="sex": This adds **color** to the points based on the sex (gender) variable, so different colors will represent males and females.

- style="time": This modifies the **shape** of the points based on the time (e.g., whether it's lunch or dinner time).


```python
# Enhanced scatter plot using seaborn
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs Tip by Gender and Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Enhanced scatter plot using seaborn
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex", style="time")
plt.title("Total Bill vs Tip by Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Step 5: Color bar

```python
# Scatter plot
## cmap is color map which can be ( "viridis", "plasma", "rainbow", "nipy_spectral" or "cividis")
sc = plt.scatter(x=data["total_bill"], y=data["tip"], c=data["total_bill"], cmap="nipy_spectral")

# Add labels
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)

# Add color bar
plt.colorbar(sc, label='Total Bill ($)')

plt.title("Total Bill vs Tip (Colored by Total Bill)")
plt.tight_layout()
plt.show()
```

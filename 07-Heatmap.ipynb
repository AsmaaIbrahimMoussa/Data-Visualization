# Heatmap
A heatmap is a visual representation of data in matrix form, where values are shown using colors.
In data analysis, it's often used to visualize correlations between numeric features.

## Step 1: Import Libraries 

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

## Step 2: Load the data

```python
data = sns.load_dataset('mpg')
data.head()
```

This dataset contains information about car fuel efficiency, such as:

- mpg: miles per gallon (fuel efficiency)
- cylinders: number of engine cylinders
- horsepower
- weight
- acceleration


## Step 3: correlation heatmap

```python
# get correlations between numerical columns
corr = data.corr(numeric_only=True)  
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True ,cmap='RdBu', vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Car Features")
plt.show()
```

Values range from -1 to 1
- 1.0: perfect positive correlation
- 1.0: perfect negative correlation
- 0: no correlation

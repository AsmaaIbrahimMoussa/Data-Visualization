# Histogram 
Histogram aims to group data into bins instead of plotting each individual value

## Step 1: Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Step 2: Load the Data

```python
data=pd.read_csv('building_data.csv')
```

```python
data.head()
```

## Step 3: Create Sipmle Histogram
Histograms help us understand how values are distributed across ranges (bins).

```python
plt.hist(data['Year of sale'])
plt.show()
```

```python
plt.hist(data['Year of sale'], bins=30)
plt.show()
```

```python
plt.hist(data['Year of sale'], bins=40, edgecolor='black')
plt.show()
```

## Step 4: Create Advanced Histogram
Weâ€™ll add color, transparency, and labels for better readability.

```python
plt.hist(data['Year of sale'], bins=30, color='red', alpha=0.5 , edgecolor='black')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(2003,2009)
plt.savefig('histogram.png', dpi=300)
plt.show()
```

```python
# Calculate statistics
mean = data['Month of sale'].mean()
median = data['Month of sale'].median()
mode = data['Month of sale'].mode()[0]  

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Plot histogram
data['Month of sale'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Add lines for mean, median, and mode
plt.axvline(mean, color='blue', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median, color='green', linestyle='dashed', linewidth=7, label='Median')
plt.axvline(mode, color='red', linestyle='dashed', linewidth=5, label='Mode')

# Show legend and plot
plt.legend()
plt.tight_layout()
plt.show()
```

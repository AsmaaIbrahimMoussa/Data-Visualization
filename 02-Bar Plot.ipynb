# Bar Plot

## Step 1: Import Libraries
We start by importing the required libraries.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Step 2: Load the Data
We load the CSV file containing developer age and salary information.

```python
data = pd.read_csv("Dev_Salary.csv")
```

```python
data.head()
```

## Step 3: Basic Bar Plot
Now we plot salaries using bar plots for different developer groups.

```python
Dev_age = data["Age"]
Dev_salary = data["All_Devs"]
plt.bar(Dev_age, Dev_salary, label='All Devs', color='b')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.title('Median Salary vs Age')
plt.legend()
plt.show()
```

```python
Dev_age = data["Age"]
Dev_salary = data["All_Devs"]
plt.bar(Dev_age , Dev_salary, color="#444444", label="All Devs")

py_dev_y = data["Python"]
plt.plot(Dev_age, py_dev_y, color="#008fd5", label="Python")

js_dev_y = data["JavaScript"]
plt.plot(Dev_age, js_dev_y, color="#e5ae38", label="JavaScript")

plt.legend()
plt.title('Median Salary vs Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

## Step 4: Customize the Plot
We enhance the plot with colors, legend, and spacing.

```python
import numpy as np

Dev_age = data["Age"]
x_indexes = np.arange(len(Dev_age))
width = 0.25

plt.figure(figsize=(20, 10))

Dev_salary = data["All_Devs"]
plt.bar(x_indexes-width , Dev_salary, width=width ,color="#444444", label="All Devs")

py_dev_salary = data["Python"]
plt.bar(x_indexes, py_dev_y, width=width ,color="#008fd5", label="Python")

js_dev_salary = data["JavaScript"]
plt.bar(x_indexes+width, js_dev_y, width=width ,  color="#e5ae38", label="JavaScript")


plt.title('Median Salary vs Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```
